```python
import torch
import random
```

```python
class DynamicNet(torch.nn.Module):
    def __init__(self, D_in, H, D_out):
        super(DynamicNet, self).__init__()
        self.input_layer = torch.nn.Linear(D_in, H)
        self.middle_layer = torch.nn.Linear(H, H)
        self.output_layer = torch.nn.Linear(H, D_out)
        
    def forward(self, x):
        h_relu = self.input_layer(x).clamp(min=0)
        for _ in range(random.randint(0,3)):
            h_relu = self.middle_layer(h_relu).clamp(min=0)
        output = self.output_layer(h_relu)
        return output
```

```python
net = DynamicNet(1000, 100, 10)

x = torch.randn(64, 1000)
y = torch.randn(64, 10)

criterion = torch.nn.MSELoss(reduce=True)
optimizer = torch.optim.SGD(params=net.parameters(), lr=1e-4, momentum=0.9)
```

```python
for i in range(500):
    y_pred = net(x)
    loss = criterion(y_pred, y)
    
    print(i, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
```
